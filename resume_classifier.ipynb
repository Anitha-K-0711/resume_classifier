# NLP steps to convert resume content to simplified text

```python
# Replacing punctuations with space
df['resume_processed'] = df['resume'].str.replace("[^a-zA-Z0-9]", " ")
# replacing all the other symbols except a-z A-Z 0-9 with a blank space
```

```python
df['resume_processed'] = df['resume_processed'].apply(lambda row: ' '.join([word for word in row.split() if len(word)>2]))
# this will delete the short 2 letter words that were there between the html tags, and small words like is, in, if, a, etc
```

```python
# converting it to lowercase characters
df['resume_processed'] = [word.lower() for word in df['resume_processed']]
```

```python
# stop words removal
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk import word_tokenize

stop_words = stopwords.words('english')

def remove_sw(resume):
  resume_tokenized = word_tokenize(resume)
  resume_new = " ".join([i for i in resume_tokenized if i not in stop_words])
  return resume_new

df['resume_processed'] = [remove_sw(word) for word in df['resume_processed']]
```

```python
df['resume_processed']
```

```python
# lemmatization
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()

# converting detailed pos to shallow pos
def nltk_tag_to_wordnet_tag(nltk_tag):
  if nltk_tag.startswith('J'):
    return wordnet.ADJ
  elif nltk_tag.startswith('V'):
    return wordnet.VERB
  elif nltk_tag.startswith('N'):
    return wordnet.NOUN
  elif nltk_tag.startswith('R'):
    return wordnet.ADV
  else:
    return None

def lemmatize_sentence(sentence):
  # word tokenize -> pos tag (detailed) -> wordnet tag (shallow pos) -> lemmatizer -> root word
  nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
  wordnet_tagged = map(lambda x: (x[0], nltk_tag_to_wordnet_tag(x[1])), nltk_tagged) # this map function goes through the list of
  # words stored in nltk_tagged in the form of tuples. # the x[0] in the tuple is the word. It takes the word same as the word
  # the x[1] is the pos of the word where the nltk_tag_to_wordnet_tag() funtion is applied to it

  lemmatized_sentence = []
  for word,tag in wordnet_tagged:
    if tag is None:
      lemmatized_sentence.append(word)
    else:
      lemmatized_sentence.append(lemmatizer.lemmatize(word,tag))
  return " ".join(lemmatized_sentence)

df['resume_processed'] = df['resume_processed'].apply(lambda x: lemmatize_sentence(x))
```

# Fit XGBoost model on 1-gram TFIDF vectorized data

```python
# Bag of words model with 1-gram tf-idf approach
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features = 3000)

x = tfidf.fit_transform(df['resume_processed']).toarray()
y = df['category'].values

feature_names = tfidf.get_feature_names_out()

from sklearn.model_selection import train_test_split
x_tr, x_te, y_tr, y_te = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score

model = xgb.XGBClassifier()
model.fit(x_tr,y_tr)

y_pred = model.predict(x_te)

```

```python
y_pred_proba = model.predict_proba(x_te)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_te,y_pred_proba, multi_class="ovr")
```

```python
# 0.96 is the AUROC score which is a very good score. So, the model is a good and descent model to classify the resumes
```
